```python
# 1.2 Изучите распределение стоимости недвижимости
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['price'], bins=50, kde=True, color='blue')
plt.title("1.2 Распределение стоимости недвижимости", fontsize=16)
plt.xlabel("Цена (USD)", fontsize=12)
plt.ylabel("Частота", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("1.2 Вывод: Цены на недвижимость имеют широкий диапазон, с пиком в области низких цен (до $500,000).")

# 1.3 Изучите распределение жилой площади
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['sqft_living'], bins=50, kde=True, color='green')
plt.title("1.3 Распределение жилой площади", fontsize=16)
plt.xlabel("Площадь (кв. футы)", fontsize=12)
plt.ylabel("Частота", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("1.3 Вывод: Жилая площадь большинства объектов недвижимости находится в диапазоне от 1000 до 3000 квадратных футов.")

# 1.4 Изучите распределение года постройки
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['yr_built'], bins=50, kde=True, color='orange')
plt.title("1.4 Распределение года постройки", fontsize=16)
plt.xlabel("Год постройки", fontsize=12)
plt.ylabel("Частота", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("1.4 Вывод: Большинство объектов недвижимости построено в период с 1950 по 2000 годы.")
```

```python
# 2.1 Изучите распределение домов от наличия вида на набережную
plt.figure(figsize=(8, 6))
waterfront_counts = kc_house_data['waterfront'].value_counts()
sns.barplot(x=waterfront_counts.index, y=waterfront_counts.values, color="skyblue")
plt.title("Распределение домов от наличия вида на набережную", fontsize=16)
plt.xlabel("Вид на набережную (0 = Нет, 1 = Да)", fontsize=12)
plt.ylabel("Количество домов", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("2.1 Вывод: Большинство домов не имеют вида на набережную. Дома с видом на набережную составляют меньшинство.")

# 2.2 Изучите распределение этажей домов
plt.figure(figsize=(8, 6))
sns.histplot(kc_house_data['floors'], bins=10, kde=False, color="purple")
plt.title("Распределение этажей домов", fontsize=16)
plt.xlabel("Количество этажей", fontsize=12)
plt.ylabel("Количество домов", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("2.2 Вывод: Большинство домов имеют 1 или 2 этажа. Дома с большим количеством этажей встречаются редко.")

# 2.3 Изучите распределение состояния домов
plt.figure(figsize=(8, 6))
condition_counts = kc_house_data['condition'].value_counts()
sns.barplot(x=condition_counts.index, y=condition_counts.values, color="lightgreen")
plt.title("Распределение состояния домов", fontsize=16)
plt.xlabel("Состояние дома (1-5)", fontsize=12)
plt.ylabel("Количество домов", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("2.3 Вывод: Большинство домов находятся в состоянии 3 или 4. Ветхие дома (1 или 2) встречаются редко.")
```

# Условие 3: 3 задача
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
# 3.1 Влияние количества спален на стоимость
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=kc_house_data, color="lightblue")
plt.title("Влияние количества спален на стоимость недвижимости", fontsize=16)
plt.xlabel("Количество спален", fontsize=12)
plt.ylabel("Стоимость недвижимости (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("3.1 Вывод: Увеличение количества спален в целом связано с ростом стоимости. Однако дома с 10 и более спальнями имеют меньшую медианную стоимость, возможно из-за их меньшей популярности.")

# 3.2 Влияние жилой площади на стоимость
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=kc_house_data, alpha=0.6, color='green', edgecolor=None)
plt.title("Влияние жилой площади на стоимость недвижимости", fontsize=16)
plt.xlabel("Жилая площадь (кв. футы)", fontsize=12)
plt.ylabel("Стоимость недвижимости (USD)", fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("3.2 Вывод: Стоимость недвижимости прямо пропорциональна жилой площади. Однако у больших объектов есть точки с низкой стоимостью, что может указывать на старые или плохо расположенные дома.")

# 3.3 Влияние состояния дома на стоимость
plt.figure(figsize=(8, 6))
sns.barplot(x='condition', y='price', data=kc_house_data, errorbar=None, color="lightgreen")
plt.title("Влияние состояния дома на стоимость недвижимости", fontsize=16)
plt.xlabel("Состояние дома (1-5)", fontsize=12)
plt.ylabel("Средняя стоимость недвижимости (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("3.3 Вывод: Состояние дома влияет на его стоимость. Дома с состоянием 5 (лучшее) имеют значительно более высокую стоимость, чем дома с состоянием 1.")

# 3.4 Влияние количества этажей на стоимость
plt.figure(figsize=(8, 6))
sns.boxplot(x='floors', y='price', data=kc_house_data, color="purple")
plt.title("Влияние количества этажей на стоимость недвижимости", fontsize=16)
plt.xlabel("Количество этажей", fontsize=12)
plt.ylabel("Стоимость недвижимости (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("3.4 Вывод: Дома с 1 или 2 этажами имеют более низкую медианную стоимость, чем дома с большим количеством этажей.")

# 3.5 Влияние вида на набережную на стоимость
plt.figure(figsize=(8, 6))
sns.boxplot(x='waterfront', y='price', data=kc_house_data, color="orange")
plt.title("Влияние вида на набережную на стоимость недвижимости", fontsize=16)
plt.xlabel("Вид на набережную (0 = Нет, 1 = Да)", fontsize=12)
plt.ylabel("Стоимость недвижимости (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("3.5 Вывод: Наличие вида на набережную оказывает значительное влияние на стоимость недвижимости. Дома с видом на набережную значительно дороже.")
```
